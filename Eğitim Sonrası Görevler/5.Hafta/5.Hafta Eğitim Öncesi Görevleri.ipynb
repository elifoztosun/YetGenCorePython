# Problem 2
Kendisine gönderilen 2 sayı arasındaki tüm asal sayıları bulan fonksiyonu yazınız.

````python
lowerBoundary = int(input())
upperBoundary = int(input())

primeNumbers = []

for num in range(lowerBoundary, upperBoundary):
    isPrime = True
    if num <= 1:
        continue
    for i in range(2, num):
        if num % i == 0:
            isPrime = False
            break
    if isPrime:
        primeNumbers.append(num)
        
print(primeNumbers)
````
