```python

class Kisi:
    def __init__(self, isim_soyisim, yas, boy, kilo):
        self.isim_soyisim = isim_soyisim
        self.yas = yas
        self.boy = boy
        self.kilo = kilo
        self.vki = self.vki_hesapla()
        self.durum = self.durum_belirle()

    def vki_hesapla(self):
        return self.kilo / (self.boy / 100) ** 2

    def durum_belirle(self):
        if self.vki < 18.5:
            return "Zayıf"
        elif 18.5 <= self.vki < 25:
            return "Normal kilolu"
        elif 25 <= self.vki < 30:
            return "Fazla kilolu"
        elif 30 <= self.vki < 40:
            return "Obez"
        else:
            return "İleri derecede obez (morbid obez)"
zayif= []
normal=[]
fazlaKilolu=[]
obez=[]
ileriObez=[]
def veri_girisi():
    kisi_sayisi = int(input("Kaç giriş yapacaksınız? "))
    global zayif
    global normal
    global fazlaKilolu
    global obez
    global ileriObez

    for _ in range(kisi_sayisi):
        isim_soyisim = input("İsim Soyisim: ")
        yas = int(input("Yaş: "))
        boy = float(input("Boy (cm): "))
        kilo = float(input("Kilo (kg): "))
        
        kisi = Kisi(isim_soyisim, yas, boy, kilo)
        if kisi.durum=="Zayıf":
            zayif.append(kisi)
        elif kisi.durum=="Normal kilolu":
            normal.append(kisi)
        elif kisi.durum=="Fazla Kilolu":
            fazlaKilolu.append(kisi)
        elif kisi.durum=="Obez":
            obez.append(kisi)
        else:
            ileriObez.append(kisi)
            
    

def verileri_dosyaya_kaydet():
    with open("vki_liste.txt", "w", encoding="utf-8") as dosya:
        for kisi in zayif:
            dosya.write(f"Zayıflar:\n İsim Soyisim: {kisi.isim_soyisim}, Yaş: {kisi.yas}, Boy: {kisi.boy} cm, Kilo: {kisi.kilo} kg, VKİ: {kisi.vki:.2f}, Durum: {kisi.durum}\n")
        for kisi in normal :
            dosya.write(f"Normaller:\n İsim Soyisim: {kisi.isim_soyisim}, Yaş: {kisi.yas}, Boy: {kisi.boy} cm, Kilo: {kisi.kilo} kg, VKİ: {kisi.vki:.2f}, Durum: {kisi.durum}\n")
        for kisi in fazlaKilolu:
            dosya.write(f"Fazla Kilolular:\n İsim Soyisim: {kisi.isim_soyisim}, Yaş: {kisi.yas}, Boy: {kisi.boy} cm, Kilo: {kisi.kilo} kg, VKİ: {kisi.vki:.2f}, Durum: {kisi.durum}\n")
        for kisi in obez:
            dosya.write(f"Obezler:\n İsim Soyisim: {kisi.isim_soyisim}, Yaş: {kisi.yas}, Boy: {kisi.boy} cm, Kilo: {kisi.kilo} kg, VKİ: {kisi.vki:.2f}, Durum: {kisi.durum}\n")
        for kisi in ileriObez:
            dosya.write(f"İleri Obezler:\n İsim Soyisim: {kisi.isim_soyisim}, Yaş: {kisi.yas}, Boy: {kisi.boy} cm, Kilo: {kisi.kilo} kg, VKİ: {kisi.vki:.2f}, Durum: {kisi.durum}\n")
def main():
    veri_girisi()
    verileri_dosyaya_kaydet()
    print("Veriler dosyaya kaydedildi.")

if __name__ == "__main__":
    main()
```

```python
import datetime
import json

products= {
    "apple": 5,
    "blueberry": 6,
    "peach": 5,
    "strawberry": 6,
    "watermelon":4,
    "orange": 2
}    

def open_menu():
    print("Welcome to YetBar- Here is the menu, you can choose any fruits to make your special smoothie!")
    for item, price in products.items():
        print(f"{item}:{price} TL")
chosenList=[]
chosen_price=[]
totalPrice=0
def chooseProduct(): # its also calculating
    global chosenList,chosen_price, totalPrice
    value= int(input("how many fruits would you like to add?"))
    for i in range(value):
        chosen= input("please write the name of fruit ")
        new= products.get(chosen,0)
        if new!=0:
            chosenList.append(chosen)
            chosen_price.append(new)
        else:
            print(f"sorry, the menu does not include {chosen}")
    totalPrice= sum(chosen_price)
    print(f"you have chosen:{chosenList}\ntotal price: {totalPrice}TL")  
sales_log=[]
def add_sale():
    global totalPrice,sales_log
    sale= {
        "chosen fruits": chosenList,
        "price": totalPrice,
        "date": datetime.datetime.now().isoformat()
    }
    sales_log.append(sale)
    print("Sale saved succesfully")

def save_sales_log(file_path="sales_log.json"):
    with open(file_path, "a", encoding="utf-8") as file:
        json.dump(sales_log, file, ensure_ascii=False, indent=4)
    print("Sale log saved successfully.")
# kayıt sonucu biraz ilginç oldu ama gayet iyi iş bence
open_menu()
chooseProduct()
add_sale()
save_sales_log() 
```
