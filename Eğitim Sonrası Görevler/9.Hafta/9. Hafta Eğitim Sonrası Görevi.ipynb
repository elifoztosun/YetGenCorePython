Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız 
# kanal listesi:

import random
import time

kanal_liste= ["Fox TV", "Disney Chanel", "TV8", "Kanal D"]
class Kumanda:
    def __init__(self, tv_durum="kapalı", kanal_liste, ses= 0, kanal=" " ):
        self.tv_durum= tv_durum
        self.ses= ses
        self.kanal=kanal
        self.kanal_liste=kanal_liste== ["Fox TV", "Disney Chanel", "TV8", "Kanal D"]
    def tv_ac(self):
        if self.tv_durum=="kapalı":
            self.tv_durum="açık"
            print("Televizyon açıldı")
        else: 
            print("TV zaten açık")
    
    def tv_kapa(self):
        if self.tv_durum=="açık":
            self.tv_durum="kapalı"
            print("Televizyon kapatıldı.")
        else:
            print("TV zaten kapalı.")

    def ses_ayarla(self, yeni_ses):
        if 0<=yeni_ses<=30:
            self.ses=yeni_ses
            print(f"Ses seviyesi {yeni_ses} olarak ayarlandı.")
        else:
            print(f"Ses seviyesi 0 ve 30 arasında olmalıdır,tekrar deneyin.")
    
    def kanal_ekle(self,kanal):

        if kanal in self.kanal_liste:
            print(f"{kanal} zaten listeye ekli.")
        else:
            self.kanal_liste.append(kanal)
            print(f"{kanal} kanalı listeye eklendi.")

    def kanal_sayisi(self):
        print(f" toplam kanal sayısı:{len(self.kanal_liste)}")

    def random_kanal(self):
        self.kanal=random.choice(self.kanal_liste)
        print(f"rastgele seçilen kanal:{self.kanal}")

    def bilgiGor(self):
        print(f"TV Bilgileri:\n Tv durumu: {self.tv_durum}\n Ses seviyesi: {self.ses}\n Kanal Listesi: {self.kanal_liste} ")
print("""
    1. Tv Aç
    2. Tv Kapat
    3. Ses Ayarları
    4. Kanal Ekle
    5. Kanal Sayısını Öğrenme
    6. Rastgele Kanala Geçme
    7. Televizyon Bilgileri
    çıkmak için "q" ya tıklayın.
      """ )

kumanda=Kumanda()


while True:
    islem= input("yapacağınız işlemi girin")
    if islem=="q":
        print("kumanda programı sonlandı.")
        break
    if islem=="1":
        kumanda.tv_ac()
        
    elif islem=="2":
        kumanda.tv_kapa()
    elif islem=="3":
        yeni_ses= int(input("ses seviyesi girin (0-30)"))
        kumanda.ses_ayarla(yeni_ses)
    elif islem=="4":
        kanal= input("eklemek istediğiniz kanalın ismi:")
        kumanda.kanal_ekle(kanal)
    elif islem=="5":
        kumanda.kanal_sayisi()
    elif islem=="6":
        kumanda.random_kanal()
    elif islem=="7":
        kumanda.bilgiGor()
    else:
        print("Geçersiz işlem tuşladınız.")
    
```
