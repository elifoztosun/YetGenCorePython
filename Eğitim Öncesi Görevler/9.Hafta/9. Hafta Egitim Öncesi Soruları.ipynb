# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self,name,max_speed,kilometre):
        self.name=name
        self.max_speed= max_speed
        self.kilometre=kilometre

class Bus(Vehicle):
    def __init__(self, name, max_speed, kilometre, seating_capacity):
        super().__init__(name, max_speed, kilometre)
        self.seating_capacity= seating_capacity
        print(f"the seating capacity of a bus is {seating_capacity} passengers.")

buS= Bus("mercedes", 180, 20000, 50)
````
